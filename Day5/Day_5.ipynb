Tas 1  kmeans

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Generate random dataset
np.random.seed(42)
X = np.random.rand(100, 2) * 10

# Define KMeans model
kmeans = KMeans(n_clusters=3, random_state=42)

# Fit the model to the dataset
kmeans.fit(X)

# Get cluster labels and centroids
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Plot the clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', marker='o', edgecolors='k')
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='X', s=200, label="Centroids")
plt.title("K-Means Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

task 2 centroid seccion

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Step 1: Generate a random dataset
np.random.seed(42)
X = np.random.rand(100, 2) * 10  # 100 points in 2D space

# Different centroid initializations
init_methods = ['random', 'k-means++']

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

for i, init in enumerate(init_methods):
    # Step 2: Apply KMeans with different initializations
    kmeans = KMeans(n_clusters=3, init=init, random_state=42)
    kmeans.fit(X)

    # Get cluster labels and centroids
    labels = kmeans.labels_
    centroids = kmeans.cluster_centers_

    # Step 3: Plot clusters
    axes[i].scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', edgecolors='k')
    axes[i].scatter(centroids[:, 0], centroids[:, 1], c='red', marker='X', s=200, label="Centroids")
    axes[i].set_title(f"K-Means Clustering with {init} Initialization")
    axes[i].set_xlabel("Feature 1")
    axes[i].set_ylabel("Feature 2")
    axes[i].legend()

plt.tight_layout()
plt.show()
```

Task 4: Elbow Method

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
```

```python
, _ = make_blobs(n_sampls=300, centers=5, cluster_std=1.05, random_state=42)
```

```python
# Step 1: Generate a synthetic dataset
X, _ = make_blobs(n_samples=300, centers=5, cluster_std=1.05, random_state=42)

# Step 2: Standardize the dataset
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 3: Compute Within-Cluster Sum of Squares (WCSS) for different K values
wcss = []
K_values = range(1, 11)

for k in K_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)  # inertia_ gives the sum of squared distances

# Step 4: Plot the elbow curve
plt.figure(figsize=(8, 5))
plt.plot(K_values, wcss, marker='o', linestyle='-', color='b')
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Within-Cluster Sum of Squares (WCSS)")
plt.title("Elbow Method for Optimal K")
plt.xticks(K_values)
plt.grid()
plt.show()
```

Task 8: Anomaly Detection with K-means

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
np.random.seed(42)
normal_data = np.random.randn(100, 2) * 0.75
anomaly_data = np.random.randn(5, 2) * 4 + 10
X = np.vstack((normal_data, anomaly_data))
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X_scaled)
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

```python
distances = np.linalg.norm(X_scaled - centroids[kmeans.labels_], axis=1)
```

```python
threshold = np.percentile(distances, 95)
anomalies = X[distances > threshold]
```

```python
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', alpha=0.6)
plt.scatter(anomalies[:, 0], anomalies[:, 1], color='red', label='Anomalies', edgecolors='black')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color='blue', marker='X', s=200, label='Centroids')
plt.legend()
plt.title("K-Means Anomaly Detection")
plt.show()
```

```python
print("Detected Anomalies:\n", anomalies)
```

Task 9: Basic Neural Network

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
from tensorflow import keras
```

```python
X, y = make_moons(n_samples=500, noise=0.2, random_state=42)
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
model = keras.Sequential([
    keras.layers.Dense(8, activation='relu', input_shape=(2,)),
    keras.layers.Dense(1, activation='sigmoid')
])
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train, epochs=50, batch_size=16, validation_data=(X_test, y_test))
```

```python
loss, accuracy = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {accuracy:.4f}")
```
